```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./files/airfoil_self_noise.dat', sep='\t', header=None)
df.head()
```

```python
X = df[[0,1,2,3,4]].values
Y = df[5].values
```

```python
#split data set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33)
```

```python
#import and instantiate model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# train model
model.fit(X_train, y_train)
```

```python
#Evaluate the trained model on the training data
model.score(X_train, y_train)
```

```python
#Evaluate the trained model on the new data (R^2)
model.score(X_test, y_test)
```

```python
prediction = model.predict(X_test)
prediction
```

```python
#the linear regression model didnt perform so well. we try using an ensemble method
from sklearn.ensemble import RandomForestRegressor
model2 = RandomForestRegressor()
```

```python
#train
model2.fit(X_train, y_train)
```

```python
#evaluate on seen data
model2.score(X_train, y_train)
```

```python
#evaluate on new data
model2.score(X_test, y_test)
#RandomForestRegressor performs better
```

```python
#new prediction
prediction2 = model2.predict(X_test)
```
