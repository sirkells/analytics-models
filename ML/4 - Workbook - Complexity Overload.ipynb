## I. Decision Trees: A New Paradigm

```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style('darkgrid')

from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv('simulated_data.csv')
#df.rename(columns={'x (input variable)': 'x', 'y (target variable)': 'y'}, inplace=True)
df.head()
#df.to_csv('simulated_data.csv', index=False)
```

```python
# helper function
def fit_and_plot_model(df, model):
    #seperate input and target
    features = df.drop('y', axis=1)
    target = df.y
    
    #fit model (train model)
    model.fit(features, target)
    
    #plot dataset
    plt.scatter(df.x, df.y)
    
    #overlay the predicted values with the features
    plt.plot(df.x, model.predict(features), 'k--')
    
    return model
```

```python
# unconstrained DecisionTree by default
model = fit_and_plot_model(df, DecisionTreeRegressor())
```

```python
#the model is overfit
# check the difference btw the original and predicted target value
actual_y = df.y

predicted_y = model.predict(df.drop('y', axis=1))

# Total difference between actual and predicted values of y
np.sum( np.absolute( actual_y - predicted_y ) )

    
```

As you can see, our total absolute prediction error was 0. Again, because this is not on an "unseen" test set, a prediction error of 0 is a cause for concern and NOT a cause for celebration. The model is most likely overfit to the training dataset, and we'll see exactly how to validate this in the next module.
Also, note that we took the sum of the absolute values of the differences in case some differences were negative and others were positive.
In summary, the unconstrained decision tree is overfit.
Because the decision tree was unconstrained, it was allowed to make as many splits as it wanted to.
As a result, it just made branches until each x value was in its own leaf!


## II. Understanding Tree Structure

# So why did the unconstrained decision tree overfit the dataset? Well, it all comes down to the structure of a decision tree. Because a decision tree is based on creating splits / branches (instead of learning... say... a coefficient), it can theoretically have an infinite number of splits.

Practically, this means that an unconstrained decision tree will keep splitting until every single observation from the training set ends up in its own terminal leaf. And for decision trees, every single branch / split location is a model parameter, which is why decision trees are considered a complex type of model.

Let's peel back and explore the structure of our model.

First, we can visualize the tree with the graphviz library.
You might not have this library installed yet. Please check the additional resources for this module for installation instructions!

```python
# Exports the tree structure
from sklearn.tree import export_graphviz

# Allows us to render the diagram in Jupyter notebook
from graphviz import Source
```

# Next, let's save the structure as a PNG file and display it here.
After running this code, you can find the PNG file saved in the same folder as this notebook. You can open that file and zoom in to see the tree more clearly.

```python
from IPython.display import Image

def plot_and_save_decision_tree(model, out_filename):
    # Export tree structure and save to Source object
    graph = Source( export_graphviz(model, out_file=None) )

    # Pipe to PNG bytes
    tree_png = graph.pipe(format='png')

    # Write to PNG file
    with open(out_filename,'wb') as f:
        f.write(tree_png)

    # Display image
    return Image(tree_png)
```

```python
plot_and_save_decision_tree(model, 'tree_graph.png')
#graph = Source( export_graphviz(model, out_file=None) )
#a = Image(graph.pipe(format='png'))
#a
```

## III. Constrained Decision Trees

The first solution to overfitting decision trees is to constrain the decision trees. Remember, the reason the unconstrained decision tree becomes overfit is because it's allowed to keep branching willy nilly until it ends up with one leaf per observation.

There are different ways to constrain a decision tree...

The first common way is to constrain the maximum depth allowed for the tree.
In other words, we can constrain the maximum number of levels the tree is allowed to branch down. So if we set a max depth of 3, it means the tree is only allowed to branch down 3 times before it MUST arrive at a prediction.

```python
# Depth-Constrained Decision Tree
model2 = fit_and_plot_model(df, DecisionTreeRegressor(max_depth=3))
```

Wow! The difference appears to be night and day. Whereas the unconstrained decision tree was completely overfit, the constrained decision tree by depth appears to approximate the sine wave fairly well.

Also, the above chart shows the branching nature of the tree much more apparently. Notice how the predictions appear to follow a step-like pattern instead of the smooth curvy pattern from the polynomial regressions.

Next, let's display the actual tree structure.
As you can see, the maximum depth is 3.

```python
plot_and_save_decision_tree(model2, 'tree_graph2.png')
```

## IV. Random Forests


# Another common way to constrain decision trees is to set a mininum number of samples that each leaf must have.

This will guarantee a mininum number of observations in each terminal node and avoid the situation where each leaf only has one observation.


```python
# Leaf-Size-Constrained Decision Tree
model3 = fit_and_plot_model(df, DecisionTreeRegressor(min_samples_leaf=5))
```

# Next, let's display the actual tree structure.
As you can see, each leaf will have at least 5 samples.
Notice how this tree is constrained in a fundamentally different way than the depth-constrained one. This tree is allowed to have as many branches as needed as long as each leaf still has at least 5 samples.

```python
plot_and_save_decision_tree(model3, 'tree_graph3.png')
```

# Now the million dollar question is: how in the world should you choose these "structural settings" (i.e. maximum depth or minimum leaf size)?
As you might have noticed, we needed to set these structural settings BEFORE fitting the model. That means the algorithm did not learn this directly from the data. Therefore, these are called hyperparameters instead of standard model parameters.
The process of selecting the optimal hyperparameters is called tuning, and it's integral to applied machine learning. We'll explore this process in the next cornerstone project.

Within each **non-leaf node**, the first value indicates the split. For example, in the tree above, the very first split occurs at $x \leq 3.332$. In other words, you would follow the left branch if your observation were less than or equal to 3.332.

Within each **leaf**, the last value indicates the predicted value based on the decision path leading up to that leaf.

#### For example, let's say you had an observation $x = 3.5$ and you wanted to predict its corresponding $y$ value, or $\hat{y} = f(x) = f(3.5)$ where $f(x)$ represents the decision tree from Exercise 4.1 Part (C). Here's how that would work:

## <span style="color:RoyalBlue">Exercise 4.3 - Input Dimensions</span>

After you train a model with Scikit-Learn, if you want to use it to predict new observations/samples, you'll need to remember that the model will accept EITHER a **Pandas DataFrame** OR a **2-Dimensional NumPy Array**. (In fact, under the hood, Scikit-Learn actually converts a Pandas DataFrame into a 2-Dimensional NumPy Array first!)

The key word there is **2-Dimensional**. The first dimension represents the "rows" or individual observations of your dataset. The second dimension represents the "columns" or input features of your dataset. Even if you only have one input feature (like we do in this simulation), the model will still expect a 2-dimensional array.

#### For example...
* If you have 1 observation and 1 input feature, you'll need a $1 \times 1$ array.
* If you have 100 observations and 1 input feature, you'll need a $100 \times 1$ array.
* If you have 1 observation and 5 input features, you'll need a $1 \times 5$ array.
* If you have 100 observation and 5 input features, you'll need a $100 \times 5$ array.
